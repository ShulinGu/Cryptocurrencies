### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
# YOUR CODE HERE
file = "Resources/crypto_data.csv"
df = pd.read_csv(file)
df.set_index('Unnamed: 0', inplace=True)
df.index.name = None
df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
# YOUR CODE HERE
df['IsTrading'] = LabelEncoder().fit_transform(df['IsTrading'])
df_traded = df[df['IsTrading']==1]
df_traded.head(10)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
# YOUR CODE HERE
df_working = df_traded[df_traded.Algorithm.notnull()]
df_working.head(10)
```

```python
# Remove the "IsTrading" column. 
# YOUR CODE HERE
df_working.drop('IsTrading', inplace=True, axis=1)
df_working.head(10)
```

```python
# Remove rows that have at least 1 null value.
# YOUR CODE HERE
df_dropnull = df_working.dropna(how='any',axis=0)
df_dropnull.head(10)
```

```python
# Keep the rows where coins are mined.
# YOUR CODE HERE
df_mined = df_dropnull[df_dropnull['TotalCoinsMined']!=0]
df_mined.head(10)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
# YOUR CODE HERE
df_name = pd.DataFrame(
     data = df_mined, columns=['CoinName']
)
df_name.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
# YOUR CODE HERE
df_dropname = df_mined.drop('CoinName', 1)
df_dropname.head(10)
```

```python
# Use get_dummies() to create variables for text features.
# YOUR CODE HERE
X = pd.get_dummies(df_dropname)
X.head(10)
```

```python
# Standardize the data with StandardScaler().
# YOUR CODE HERE
X_scaled = StandardScaler().fit_transform(X)
X_scaled
```
